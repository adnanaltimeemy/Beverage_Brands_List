```python
import pandas as pd
import numpy as np
```

```python
# Orange soda market estimates (2025) in USD millions (proxy)
orange_soda_market = {
    "Region": ["North America", "Europe", "Asia Pacific", "Latin America", "Rest of World"],
    "Total_Orange_Soda_Sales_MUSD": [380.2, 326.0, 299.5, 86.4, 59.0]  # (millions) fictional composites
}

df_sales = pd.DataFrame(orange_soda_market)

# Fanta's reported dominant position in orange flavored drinks
# Assuming 0.50 share in each region as an example
df_sales["Fanta_Sales_MUSD"] = df_sales["Total_Orange_Soda_Sales_MUSD"] * 0.50

df_sales
```

```python
profit_scenarios = [0.20, 0.30, 0.40]

# Build DataFrame for projected profit
profit_df = pd.DataFrame({
    "Profit_Scenario": ["Low (20%)", "Medium (30%)", "High (40%)"],
    "Profit_Margin": profit_scenarios
})

# Add revenue totals
total_sales = df_sales["Fanta_Sales_MUSD"].sum()
profit_df["Estimated_Profit_MUSD"] = total_sales * profit_df["Profit_Margin"]

profit_df
```
