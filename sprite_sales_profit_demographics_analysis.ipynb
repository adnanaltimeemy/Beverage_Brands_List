```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# display dataframes more cleanly
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 30)
```

```python
# Create sales share data for Sprite vs key competitors
sales_data = pd.DataFrame({
    'Brand': ['Coca-Cola Classic', 'Sprite', 'Diet Coke', 'Pepsi', 'Other'],
    'Market_Share_%': [17.1, 8.03, 7.8, 7.97, 59.1]
})

sales_data
```

```python
plt.figure(figsize=(8,5))
plt.pie(sales_data['Market_Share_%'], labels=sales_data['Brand'], autopct='%1.1f%%', startangle=140)
plt.title("U.S. Soft Drink Market Share by Volume (Example Data)")
plt.show()
```

```python
years = [2020, 2021, 2022, 2023, 2024, 2025]
sprite_share = [6.5, 7.0, 7.5, 8.0, 8.03, 9.1]  # example trend

trend_df = pd.DataFrame({'Year': years, 'Sprite_Market_Share_%': sprite_share})
trend_df
```

```python
sns.lineplot(data=trend_df, x='Year', y='Sprite_Market_Share_%')
plt.title('Sprite Market Share Trend (20202025)')
plt.ylabel('Market Share (%)')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

```python
total_revenue_2024 = 47.1 * 1e9  # Coca-Cola revenue approx (USD)
sprite_share_2024 = 0.0803       # from sales_data
sprite_revenue_est = total_revenue_2024 * sprite_share_2024

sprite_revenue_est
```

```python
estimated_margin = 0.50
sprite_profit_est = sprite_revenue_est * estimated_margin

sprite_profit_est
```

```python
# Example synthetic demographics
demo_data = pd.DataFrame({
    'Age_Group': ['13-17','18-24','25-34','35-44','45-54','55+'],
    'Percent_of_Consumers': [12, 28, 22, 15, 13, 10]
})
demo_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))

sns.barplot(
    data=demo_data,
    x='Age_Group',
    y='Percent_of_Consumers',
    hue='Age_Group',          # Required for palette in new seaborn
    palette='viridis',
    legend=False              # Removes unnecessary legend
)

plt.title('Sprite Consumer Age Demographic (Hypothetical)')
plt.ylabel('% of Consumers')
plt.xlabel('Age Group')

plt.tight_layout()
plt.show()
```
